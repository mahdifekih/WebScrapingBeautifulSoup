```python
from bs4 import BeautifulSoup
import requests
import re # Regular Expression.
import pandas as pd 
import lxml # Used lxml as a parser
```

```python
#Using Request module to request the page's source code 
```

```python
src = "https://www.imdb.com/chart/top/"
page = requests.get(src)
```

```python
# Display the page's source code
```

```python
page.content
```

```python
soup = BeautifulSoup(page.content, 'lxml')
print(soup.prettify())
```

```python
# scrap movie names
scraped_movies = soup.find_all('td', {"class" : "titleColumn"})
print(scraped_movies)
```

```python
#parse movie names
movies = []
for movie in scraped_movies:
    movie = movie.get_text().replace('\n', "") # I used the module replace to delete (\'n')
    movie = movie.strip(" ") # I used strip modules to remove unecessary spaces from the scraped data
    movies.append(movie) 

print(movies)
```

```python
#Scrap movie ratings
scraped_ratings = soup.find_all('td', {"class":"ratingColumn imdbRating"})
scraped_ratings
```

```python
# parse ratings
ratings = []
for rating in scraped_ratings:
    rating = rating.get_text().replace('\n', '')
    ratings.append(rating)
ratings
```

```python
tag = soup.find('div', class_=re.compile(r'by+'))
print(tag.text)
```

```python
crew = [a.attrs.get('title') for a in soup.select('td.titleColumn a')]
crew
```

```python
links = ['imdb.com'+a.attrs.get('href') for a in soup.select('td.titleColumn a')]
links
```

```python
#Store the Scraped Data
```

```python
data = pd.DataFrame()
data['Movie Names'] = movies
data['Ratings'] = ratings
data['Direction and crew'] = crew
data['Link'] = links
data
```

```python
data.to_csv('IMDb Top 250 rated by IMDb voters', index=False)
```
